```python
import pandas as pd

data1 = pd.read_csv ('train.csv')
data2 = pd.read_csv ('test.csv')
data3 = pd.read_excel ('MacroEconomicDataActual.xlsx')
train1= pd.merge(data1, data3, on= ["Year","Month"], how= "inner")
test1= pd.merge(data2, data3, on= ["Year","Month"], how= "inner")
train1
```
